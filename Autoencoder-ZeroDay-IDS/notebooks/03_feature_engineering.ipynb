```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv("../dataset/processed/train_clean.csv")
test_df = pd.read_csv("../dataset/processed/test_clean.csv")

print(train_df.shape)
print(test_df.shape)
```

```python
train_df.columns
train_df.drop("id", axis=1, inplace=True)
test_df.drop("id", axis=1, inplace=True)
train_df.columns
```

```python
categorical_cols = train_df.select_dtypes(include="object").columns

print(categorical_cols)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(
    handle_unknown="use_encoded_value",
    unknown_value=-1
)

categorical_cols = ["proto", "service", "state"]

train_df[categorical_cols] = encoder.fit_transform(train_df[categorical_cols])

test_df[categorical_cols] = encoder.transform(test_df[categorical_cols])

train_df.head()
```

```python
train_df.head()
```

```python
X_train = train_df.drop(["label", "attack_cat"], axis=1)

y_train = train_df["label"]

X_test = test_df.drop(["label", "attack_cat"], axis=1)

y_test = test_df["label"]
```

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
X_train_normal = X_train[y_train == 0]
print(X_train_normal.shape)
y_train.value_counts()
print((y_train == 0).sum())
X_train_normal.shape[0]
X_train.to_csv("../dataset/processed/X_train.csv", index=False)

X_test.to_csv("../dataset/processed/X_test.csv", index=False)

X_train_normal.to_csv("../dataset/processed/X_train_normal.csv", index=False)

y_train.to_csv("../dataset/processed/y_train.csv", index=False)

y_test.to_csv("../dataset/processed/y_test.csv", index=False)
```
